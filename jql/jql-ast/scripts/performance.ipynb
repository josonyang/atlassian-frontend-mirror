```python
%matplotlib inline
```

```python
import pandas
```

```python
df = pandas.read_csv("./output/performance.csv")
```

```python
# min
df.milliseconds.min()
```

```python
# p50
df.milliseconds.quantile(.5)
```

```python
# p90
df.milliseconds.quantile(.9)
```

```python
# p95
df.milliseconds.quantile(.95)
```

```python
# p99
df.milliseconds.quantile(.99)
```

```python
# max
df.milliseconds.max()
```

```python
# histogram
histogram = df.plot.hist(
    title=False,
    by="milliseconds",
    bins=25,
    figsize=(40,30),
    grid=False,
    rwidth=.9,
    legend=False,
    fontsize=32
)

histogram.set_xlabel('Parse time (milliseconds)', fontsize=40, labelpad=40)
histogram.set_ylabel('Count', fontsize=40, labelpad=30)
```

```python
detail_histogram = df.plot.hist(
    title="Parse time (milliseconds)",
    by="milliseconds",
    bins=25,
    figsize=(40,30),
    grid=False,
    range=(0,1),
    rwidth=.9,
    legend=False,
    fontsize=32
)

detail_histogram.set_xlabel('Parse time (milliseconds)', fontsize=40, labelpad=40)
detail_histogram.set_ylabel('Count', fontsize=40, labelpad=30)
```
